```python
#Contruindo gráfico
#agrupa por classe
grupos = df_superv.groupby("Classe")

#plota
fig, ax = plt.subplots()
ax.margins(0.05)

#itera sobre cada um dos grupos
for name, grupo in grupos:
    ax.plot(grupo['Comp. Sepalas (cm)'], grupo['Larg. Sepalas (cm)'], marker='o', linestyle='', ms=12, label=name)

ax.legend()
plt.xlabel('Comp. Sepalas (cm)')
plt.ylabel('Larg. Sepalas (cm)')
plt.title('Comprimento e Largura das Sépalas')
leg = ax.legend(loc = 'lower right')
plt.show()
```

```python
#Plota scatterplot
df_n_superv.plot.scatter(x='Comp. Sepalas (cm)', y='Larg. Sepalas (cm)', marker='o')
```
