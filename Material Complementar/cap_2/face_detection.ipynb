```python
# -*- coding: utf-8 -*-    
#definição das bibliotecas  
  
import cv2  # biblioteca openCV - utilizada para os processos que envolvem o tratamento de imagem  
import matplotlib.pyplot as plt   # mostrar a imagem no tamanho desejado  
```

```python
#Realiza a classificação  
  
# Criação do objeto CascadeClassifier   
# É esse classificador que encontra a porção da imagem que corresponde às faces  
face_detection = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") #arquivo xml que contém o treinamento do HaaR cascade  https://github.com/opencv/opencv/tree/master/data/haarcascades
  
# Carregando a imagem   
img = cv2.imread("/Users/deborabastos/Documents/4.Bootcamp ML/bootcamp ML code/Material Complementar/cap_2/foto.jpeg")  # pode ser escolhida qualquer imagem para realizar a identicação  
# Lendo a imagem em uma escala de cinza  
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #carrega a imagem em escala de cinza para facilitar a classificação  
  
# procurando pelas coordenadas da imagem  
faces = face_detection.detectMultiScale(gray_img, scaleFactor = 1.05, minNeighbors=5) #  
  
for x,y,w,h in faces:       #identifica as coordenadas da foto e desenha o retângulo  
    img = cv2.rectangle(img, (x,y), (x+w,y+h),(0,255,0),3) #desenha o retângulo ao redor da imagem  
   
```

```python
#---------------------------------------------------------------------------------------------------  
# Realiza o plot das imagens  
plt.figure(figsize=(20,10)) #define o tamanho da imagem a ser plotada  
plt.imshow(img) # mostra a imagem  
   
# cv2.waitKey(0) # espera a entrada de alguma tecla para fechar a imagem criada  
   
# cv2.destroyAllWindows() #fecha a imagem  
```

```python
#Realiza a classificação  
  
# Criação do objeto CascadeClassifier   
# É esse classificador que encontra a porção da imagem que corresponde às faces  
face_detection = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") #arquivo xml que contém o treinamento do HaaR cascade  https://github.com/opencv/opencv/tree/master/data/haarcascades
eye_detection = cv2.CascadeClassifier("haarcascade_eye.xml")
  
# Carregando a imagem   
img = cv2.imread("/Users/deborabastos/Documents/4.Bootcamp ML/bootcamp ML code/Material Complementar/cap_2/foto.jpeg")  # pode ser escolhida qualquer imagem para realizar a identicação  
# Lendo a imagem em uma escala de cinza  
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #carrega a imagem em escala de cinza para facilitar a classificação  
  
# procurando pelas coordenadas da imagem  
faces = face_detection.detectMultiScale(gray_img, scaleFactor = 1.05, minNeighbors=5) #   
for x,y,w,h in faces:       #identifica as coordenadas da foto e desenha o retângulo  
    img = cv2.rectangle(img, (x,y), (x+w,y+h),(0,255,0),3) #desenha o retângulo ao redor da imagem 
    roi_gray = gray_img[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_detection.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
      cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),3)
 
```

```python
#---------------------------------------------------------------------------------------------------  
# Realiza o plot das imagens  
plt.figure(figsize=(20,10)) #define o tamanho da imagem a ser plotada  
plt.imshow(img) # mostra a imagem  
   
# cv2.waitKey(0) # espera a entrada de alguma tecla para fechar a imagem criada  
   
# cv2.destroyAllWindows() #fecha a imagem  
```
